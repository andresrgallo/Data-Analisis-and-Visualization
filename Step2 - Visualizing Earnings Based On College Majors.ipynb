```python
maj_counts = dict()
def calculate_major_cat_totals(df,strg):
    cats = df['Major_category'].unique()
    counts_dictionary = dict()
    
    for c in cats:
        major_df = df[df["Major_category"] == c]
        total = major_df[strg].sum()
        counts_dictionary[c] = total
    return counts_dictionary
maj_countsw = calculate_major_cat_totals(recent_grads,"Women")
maj_countsm = calculate_major_cat_totals(recent_grads,"Men")
print(maj_countsw)
pos = list(range(len(maj_countsw))) 
width = 0.5

fig, ax = plt.subplots(figsize=(10,5))
plt.bar(pos, maj_countsw.values(), width, color="red")
plt.bar([p + width for p in pos], maj_countsm.values(), width)
plt.xticks(range(len(maj_countsw)),maj_countsw.keys(), rotation = 90)

ax.set_ylabel('Quantity')

ax.set_title('Women vs Men per Major Categories')

ax.set_xticks([p + 1.5 * width for p in pos])
#ax.set_xticklabels(maj_countsw.keys)


plt.xlim(min(pos)-width, max(pos)+width*4)
#plt.ylim([0, max(maj_countsw['Major_category'] + maj_countsm['Major_category'])])

# Adding the legend and showing the plot
plt.legend(['Women', 'Men'], loc='upper left')
plt.grid()
plt.show()




catmen = {}
catwomen = {}
cats1 = recent_grads["Major_category"].unique()
for i in cats1:
    catrow = recent_grads[recent_grads["Major_category"] == i]
    totalm = catrow["Men"].sum()
    totalw = catrow["Women"].sum()
    catmen[i] = totalm
    catwomen[i] = totalw

catwomenshare = {}
for i in cats1:
    catrow = recent_grads[recent_grads["Major_category"] == i]
    totalm = catrow["Men"].sum()
    totalw = catrow["Women"].sum()
    catwomenshare[i] = totalw / (totalm + totalw)
   
sorted(catwomenshare.items(), key=lambda x: -x[1])


```
